# bar graph

```python
x1=np.array([5,8,10])
y1=np.array([12,6,6])
x2=np.array([6,9,10])
y2=np.array([6,15,7])
plt.bar(x1,y1,label="bar one",color='m')
plt.bar(x2,y2,label="bar two",color='c')
plt.legend()
plt.xlabel("bar number")
plt.ylabel("bar values")
plt.title("sample bar graph")
plt.show()
```

# scatter plot

```python
x1=np.array([5,8,10])
y1=np.array([12,6,6])
plt.scatter(x1,y1,label="scat",color='r')
plt.xlabel("x scat")
plt.ylabel("y scat")
plt.title("scatterplot")
plt.show()
```

# pie chart

```python
a=[10,3,4,12]
b=['green','blue','black','red']
colors=['g','b','k','r']
plt.pie(a,labels=b,colors=colors,startangle=90,shadow=True)
plt.title("pie chart")
plt.show()
```

# linear graph

```python
x=np.array([1,2,3])
y=np.array([5,7,4])
plt.plot(x,y,label="line")
plt.title("sample linear graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```
